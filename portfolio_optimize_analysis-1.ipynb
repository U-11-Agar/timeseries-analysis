portfolio optimization by minimizing MCVaR

```python
portfolio=pd.DataFrame()
weights=[1]*len(s_df.columns)
weights/=np.sum(weights)
print(dict(zip(df.columns,weights)))

portfolio['bank_nifty']=np.dot(df,weights)
```

```python
# !pip install pyportfolioopt
from pypfopt.efficient_frontier import EfficientFrontier
from pypfopt.expected_returns import mean_historical_return
from pypfopt.expected_returns import capm_return
from pypfopt.expected_returns import ema_historical_return
from pypfopt.risk_models import CovarianceShrinkage
from pypfopt.discrete_allocation import DiscreteAllocation, get_latest_prices
from pypfopt import HRPOpt
from pypfopt.efficient_frontier import EfficientCVaR
```

building portfolio with Mean - variance

```python

mu = mean_historical_return(df)
# mu = capm_return(df)
# mu= ema_historical_return(df)
S = CovarianceShrinkage(df).ledoit_wolf()


ef = EfficientFrontier(mu, S)
mv_weights = ef.max_sharpe()


print(dict(mv_weights))


ef.portfolio_performance(verbose=True)


latest_prices = get_latest_prices(df)

da = DiscreteAllocation(mv_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da.greedy_portfolio()
print("Discrete allocation:", allocation)
print("Funds remaining: ${:.2f}".format(leftover))
```

```python

returns = df.pct_change().dropna()
hrp = HRPOpt(returns)
hrp_weights = hrp.optimize()
hrp.portfolio_performance(verbose=True)
print(dict(hrp_weights))

da_hrp = DiscreteAllocation(hrp_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da_hrp.greedy_portfolio()
print("Discrete allocation (HRP):", allocation)
print("Funds remaining (HRP): ${:.2f}".format(leftover))
```

```python
S = df.cov()
ef_cvar = EfficientCVaR(mu, S)
cvar_weights = ef_cvar.min_cvar()

# cleaned_weights = ef_cvar.clean_weights()
print(dict(cvar_weights))

ef_cvar.portfolio_performance(verbose=True)

da_cvar = DiscreteAllocation(cvar_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da_cvar.greedy_portfolio()
print("Discrete allocation (CVaR):", allocation)
print("Funds remaining (CVaR): ${:.2f}".format(leftover))
```

```python
from pypfopt.efficient_frontier import EfficientCDaR
ef_cdar=EfficientCDaR(mu,S)
cdar_weights=ef_cdar.min_cdar()
print(dict(cdar_weights))

ef_cdar.portfolio_performance(verbose=True)

da_cdar = DiscreteAllocation(cdar_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da_cdar.greedy_portfolio()
print("Discrete allocation (CDaR):", allocation)
print("Funds remaining (CDaR): ${:.2f}".format(leftover))
```

```python
from pypfopt.efficient_frontier import EfficientSemivariance
ef_semivariance=EfficientSemivariance(mu,S,solver='SCS')
# ef_semivariance.solver = 'MOSEK'
semivariance_weights=ef_semivariance.min_semivariance()
print(dict(semivariance_weights))

ef_semivariance.portfolio_performance(verbose=True)

da_semivariance = DiscreteAllocation(semivariance_weights, latest_prices, total_portfolio_value=100000)

allocation, leftover = da_semivariance.greedy_portfolio()
print("Discrete allocation (CVAR):", allocation)
print("Funds remaining (CVAR): ${:.2f}".format(leftover))
```

```python
portfolio['MV']=np.dot(df,list(mv_weights.values()))
portfolio['HRP']=np.dot(df,list(hrp_weights.values()))
portfolio['CVaR']=np.dot(df,list(cvar_weights.values()))
portfolio['CDaR']=np.dot(df,list(cdar_weights.values()))
portfolio['SemiVariance']=np.dot(df,list(semivariance_weights.values()))
```
